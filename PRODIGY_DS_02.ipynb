```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Titanic-Dataset.csv')
data
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:")
print(data.head())
```

```python
data.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
```

```python
data['PassengerId'] = encoder.fit_transform(data[['PassengerId']])
data['Survived'] = encoder.fit_transform(data[['Survived']])
data['Pclass'] = encoder.fit_transform(data[['Pclass']])
data['Name'] = encoder.fit_transform(data[['Name']])
data['Sex'] = encoder.fit_transform(data[['Sex']])
data['Age'] = encoder.fit_transform(data[['Age']])
data['SibSp'] = encoder.fit_transform(data[['SibSp']])
data['Parch'] = encoder.fit_transform(data[['Parch']])
data['Ticket'] = encoder.fit_transform(data[['Ticket']])
data['Fare'] = encoder.fit_transform(data[['Fare']])
data['Cabin'] = encoder.fit_transform(data[['Cabin']])
data['Embarked'] = encoder.fit_transform(data[['Embarked']])
```

```python
data
```

```python
# 1. Check for missing values
print("\nMissing Values Summary:")
print(data.isnull().sum())
```

```python
# 3. Summary statistics of the dataset
print("\nSummary Statistics:")
print(data.describe())
```

```python
# 2. Data Cleaning: Fill missing values or drop them
# Fill missing 'Age' values with median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column since it has too many missing values
data.drop('Cabin', axis=1, inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
# 4. Exploratory Data Analysis (EDA)

# a. Distribution of passengers by survival status
sns.countplot(data=data, x='Survived', palette='Set2')
plt.title('Distribution of Survival (0 = Not Survived, 1 = Survived)')
plt.show()
```

```python
# b. Distribution of passengers by class
sns.countplot(data=data, x='Pclass', palette='Set3')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Check data types of columns
print(data.dtypes)

# Ensure 'Survived' and 'Sex' columns are correctly mapped and of the correct type
data['Survived'] = data['Survived'].astype(str)  # Convert to string if necessary
data['Sex'] = data['Sex'].astype(str)

# Gender distribution by survival
sns.countplot(data=data, x='Sex', hue='Survived', palette='Set1')
plt.title('Survival by Gender')
plt.show()
```

```python
# f. Correlation heatmap between features
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state= 1234)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10)
```

```python
model.fit(X_train,y_train)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
y_pred_prob = model.predict_proba(X_test)
y_pred_prob
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

print('accuracy of the model:',accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_test))
```

```python
sns.heatmap(confusion_matrix(y_pred,y_test), annot = True, fmt='.2f')
```
